```python
import os


def print_tree(path, depth=1, _current_depth=0, _prefix=""):
  """Prints a tree structure of directories up to the given depth."""
  if _current_depth == 0:
    print(path)

  if _current_depth >= depth:
    return

  try:
    entries = sorted(
        [e for e in os.scandir(path) if e.is_dir()],
        key=lambda e: e.name.lower()
    )
  except PermissionError:
    return

  for i, entry in enumerate(entries):
    is_last = i == len(entries) - 1
    connector = "└── " if is_last else "├── "
    print(f"{_prefix}{connector}{entry.name}")
    extension = "    " if is_last else "│   "
    print_tree(entry.path, depth, _current_depth + 1, _prefix + extension)
```
