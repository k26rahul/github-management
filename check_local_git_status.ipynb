```python
import json
import os
import subprocess
```

```python
def check_local_git_status(mapping_filename):
  with open(mapping_filename, 'r', encoding='utf-8') as f:
    mapping = json.load(f)

  clean_repos = 0
  dirty_repos = 0

  for repo_name, local_path in mapping.items():
    if not os.path.exists(local_path):
      print(f'[!] {repo_name}: path not found ({local_path})')
      continue

    try:
      result = subprocess.run(
          ["git", "status", "--short"],
          cwd=local_path,
          capture_output=True,
          text=True,
          check=True
      )
      status_output = result.stdout.strip()

      if status_output:
        dirty_repos += 1
        print(f'\n[DIRTY] {repo_name} ({local_path}):')
        print('    ' + status_output.replace('\n', '\n    '))
      else:
        clean_repos += 1

    except subprocess.CalledProcessError as e:
      print(f'[ERROR] {repo_name}: {e}')

  print(f'\nSUMMARY: {dirty_repos} dirty, {clean_repos} clean.')
```

```python
check_local_git_status("repo_path_mapping.json")
```
