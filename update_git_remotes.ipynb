```python
import json
import os
import subprocess
```

```python
def update_git_remotes(username="k26rahul", dry_run=True):
  with open('repo_path_mapping.json', 'r', encoding='utf-8') as f:
    mapping = json.load(f)

  for repo_name, local_path in mapping.items():
    new_url = f"https://github.com/{username}/{repo_name}.git"

    if not os.path.exists(local_path):
      print(f'[!] {repo_name}: path not found')
      continue

    try:
      current_url = subprocess.run(
          ["git", "remote", "get-url", "origin"],
          cwd=local_path,
          capture_output=True,
          text=True
      ).stdout.strip()

      if current_url == new_url:
        continue

      if dry_run:
        print(f'[DRY RUN] {repo_name}: {current_url} -> {new_url}')
      else:
        subprocess.run(
            ["git", "remote", "set-url", "origin", new_url],
            cwd=local_path,
            check=True,
            capture_output=True,
            text=True
        )
        print(f'[OK] {repo_name}')

    except subprocess.CalledProcessError as e:
      print(f'[ERROR] {repo_name}: {e.stderr.strip()}')
    except Exception as e:
      print(f'[ERROR] {repo_name}: {e}')
```

```python
update_git_remotes(dry_run=True)
```
